1. Factorial of a number using while loop

```python
n=5
f=1
while(n>0):
  f = f * n
  n = n - 1
print(f)
```

2. Factorial of a number using for loop

```python
n= 5
fact = 1
for i in range(n,0,-1):
  fact = fact * i
print(fact)
```

3. Check whether the given number is prime

```python
n= 11
flag = True
if n>0:
  for i in range(2,n):
    if n%i==0:
      print("Not Prime")
      flag = False
      break
  else:
    print("Prime")
```

4. Find a prime number in given range

```python
start = 10
end = 50
for i in range(start,end+1):
  for j in range(2,i):
    if i%j == 0:
      break
  else:
    print(i)
```

7. Check whether two strings are anagrams using simple apporach

    s1 = "listen"
    s2 = "silent"

```python
s1 = "listen"
s2 = "silent"

dct1 = {}
for i in range(0,len(s1)):
  c = s1.count(s1[i])
  dct1[s1[i]] = c
print(dct1)

dct2 = {}
for i in range(0,len(s2)):
  c = s2.count(s2[i])
  dct2[s2[i]] = c
print(dct2)

if dct1 == dct2:
  print("Anagram")
else:
  print("Not Anagram")
```

10. Check whether the number is armstrong number

1³ + 5³ + 3³ = 1 + 125 + 27 = 153.

```python
n=153
no = str(153)
sum1 = 0
for i in range(0,len(no)):
  x = int(no[i]) ** 3
  sum1 = sum1 + x

if sum1 == n:
  print("Armstrong")
else:
  print("Not Armstrong")
```

11. Two Sum

arr = [2, 7, 11, 15]
target = 9

two numbers whose sum = target

```python
arr = [2, 7, 11, 15]
target = 9

for i in range(0,len(arr)):
  for j in range(i+1,len(arr)):
    if arr[i]+arr[j]==target:
      print(arr[i],arr[j])
```

12. Missing number in an array

arr = [3, 0, 1]
Numbers should be: 0,1,2,3
👉 Missing = 2

Formula = n * (n + 1) // 2​

```python
arr = [3,0,1]

n = len(arr)
total = n *(n+1)//2

actual_sum = sum(arr)
print(total - actual_sum)
```
