4. Function Returing multiply values

Create a function that returns both the area and circumference of a circle given its radius.

```python
import math

def fun(radius):
  area = math.pi*radius*radius
  circumference = 2*math.pi*radius
  return round(area,2),round(circumference,2)

area, circumference = fun(2)
print("Area:",area)
print("Circumference:",circumference)
```

7. Function with *args

Write a function that takes a varibale number of arguments and return their sum.

(ex: if we give 2 numbers it should sum up them, but if we give 5 numbers also it should sum up)

```python
def sumAll(*args):
  print("args values",args)
  for i in args:
    print(i)
  return sum(args)

print(sumAll(1,2,3))
print(sumAll(1,2,3,4,5,6))
print(sumAll(1,2,3,4,5,6,7,8))
```

8. Function with *kwargs

Create a function that accepts any number of keyword arguments and print then in format key:value

**kwargs (keyword arguments) is a special syntax that allows a function to accept any number of named (keyword) arguments.

```python
def kwargs_function(**kwargs):
  for k,v in kwargs.items():
    print(k,":",v)

print(kwargs_function(name="Sakshi",age=20))
print(kwargs_function(name="Sakshi",age=20,city="Pune"))
print(kwargs_function(name="Sakshi",age=20,city="Pune",country="India"))
```

9. Generate a function with yield

Write a function that yields even numbers up to a specific limit

👉 yield is used to return a value from a function WITHOUT ending the function

👉 It turns a function into a generator

```python
def evenNos_generator(no_limt):
  for i in range(2,no_limt+1,2):
    yield i

for i in evenNos_generator(10):
  print(i)
```

10. Recursion Function

Write a program to find a factorial of a number using Recursion function

```python
def fact(n):
  if n == 0:
    return 1
  else:
    return n*fact(n-1)

fact(5)
```
