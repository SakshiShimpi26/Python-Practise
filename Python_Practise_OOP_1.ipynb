2. Class Method and Self

Add a method to the car class that display the full name of the car (ex. Brand model)

3. Inheritance

Create a ElectricCar class that inherits from the Car class and has a extra attribute battery_size

```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def full_name(self):
    self.full_name = self.brand + " " + self.model
    return self.full_name

class ElectricCar(Car):
  def __init__(self,brand,model,battery_size):
    super().__init__(brand,model)
    self.battery_size = battery_size

e1 = ElectricCar("Tesla", "Model S", 75)
print(e1.brand)
print(e1.model)
print(e1.full_name())
```

4. Encapsulation

Modify the Car class to encapsulate the brand attribute, making it private , and provide a getter method for it

```python
class Car:
  def __init__(self, brand, model):
    self.__brand = brand #adding two __ in any attribute make its private
    self.model = model

  def full_name(self):
    self.full_name = self.brand + " " + self.model
    return self.full_name

  def get_brand(self): # get_... this is getter method that can access the private attributes
    return self.__brand

c1 = Car("Toyota", "Camry")
print(c1.get_brand()) # Will print the value of brand

# But if we now directly try to access brand we will get error
# print(c1.brand)
# AttributeError: 'Car' object has no attribute 'brand'
```

5. Polymorphism

Demonstarate a polymorphism by defining a method fuel_type in both Car and ElectricCar classes, but with different behaviours.

```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def full_name(self):
    self.full_name = self.brand + " " + self.model
    return self.full_name

  def fuel_type(self):
    return "Petrol or Diesel"

class ElectricCar(Car):
  def __init__(self,brand,model,battery_size):
    super().__init__(brand,model)
    self.battery_size = battery_size

  def fuel_type(self):
    return "Electricity"

c1 = Car("Toyota", "Camry")
print(c1.fuel_type())

c2 = ElectricCar("Tesla", "Model S", 75)
print(c2.fuel_type())
```

6. Electric Car

Add a class variable to Car that keeps track of the number of cars created

```python
class Car:
  number_of_cars = 0

  def __init__(self, brand, model):
    self.brand = brand
    self.model = model
    Car.number_of_cars +=1

  def full_name(self):
    self.full_name = self.brand + " " + self.model
    return self.full_name

class ElectricCar(Car):
  def __init__(self,brand,model,battery_size):
    super().__init__(brand,model)
    self.battery_size = battery_size

c1 = Car("Toyota", "Camry")

c2 = ElectricCar("Tesla", "Model S", 75)

print(Car.number_of_cars)
```

7. Static Method

Add a static method to a class Car that returns a general description of the Car

(Explanation: A method is not accessible by the object of class but directly can be accesses using Class name)

```python
class Car:
  number_of_cars = 0

  def __init__(self, brand, model):
    self.brand = brand
    self.model = model
    Car.number_of_cars +=1

  def full_name(self):
    self.full_name = self.brand + " " + self.model
    return self.full_name

  @staticmethod
  def general_description():
    return "This is a Car"

class ElectricCar(Car):
  def __init__(self,brand,model,battery_size):
    super().__init__(brand,model)
    self.battery_size = battery_size

c1 = Car("Toyota", "Camry")

c2 = ElectricCar("Tesla", "Model S", 75)

print(Car.general_description())
```

8. Property Decorators

Use a property decorator in the Car class to make the model attribute read-only

```python
class Car:
  number_of_cars = 0

  def __init__(self, brand, model):
    self.brand = brand
    self.__model = model
    Car.number_of_cars +=1

  def full_name(self):
    self.full_name = self.brand + " " + self.__model
    return self.full_name

  @property
  def model(self):
    return self.__model

  @staticmethod
  def general_description():
    return "This is a Car"

c1 = Car("Toyota", "Camry")

c2 = ElectricCar("Tesla", "Model S", 75)

print(c1.model)
```
