1. Age Group Categorization

Classify a person's age group: Child (< 13), Teenager (13–19), Adult (20–59), Senior (60+).

```python
inp = int(input("Enter your age :"))
if inp>0 and inp<13:
  print("Child")
elif inp>=13 and inp<=19:
  print("Teenager")
elif inp>=20 and inp<=59:
  print("Adult")
elif inp>=60 and inp<=100:
  print("Senior")
else:
  print("Not Human")
```

2. Movie Ticket Pricing

Movie tickets are priced based on age : $12 for adults(18 or above), $8 for childers. Everyone gets a $2 discount on Wednesday

```python
age = int(input("Age: "))
day = input("Day: ")

if age >= 18:
  if day.lower()=="wednesday":
    price = 100 - 12
    print("Original Price = 100", "Discounted Price = ", price)
  else:
    print("Ticket Price = 100")
else:
  if day.lower()=="wednesday":
    price = 100 - 8
    print("Original Price = 100", "Discounted Price = ",price)
  else:
    print("Ticket Price = 100")
```

3. Grade Calculator

Assign a grade to student based on marks:

A = 90 to 100

B = 80 to 89

C = 70 to 79

```python
marks = int(input("Enter marks: "))
if marks>=90 and marks<=100:
  print("Grade: A")
elif marks>=80 and marks<90:
  print("Grade: B")
elif marks>=70 and marks<80:
  print("Grade: C")
elif marks>=60 and marks<70:
  print("Grade: D")
else:
  print("Grade: F")
```

4. Fruit Ripness Checker

Determine if a fruit is ripe, overripe, or unripe based on its color:

Fruit: Banana

Green --> Unripe

Yellow --> Ripe

Brown --> Overripe

```python
inp = input("Enter the color of banana:")
if inp.lower()=="green":
  print("Unripe")
elif inp.lower() =="yellow":
  print("Ripe")
elif inp.lower() == "brown":
  print("Overripe")
else:
  print("No such color exists in banana")
```

9. Leap Year Checker

Check if a year is a leap year

Leap year is divisible by 4,  but not by 100 unless also divisible  by 400.

```python
year = int(input("Enter the year: "))
if (year%400 == 0) or (year%4==0 or year%100!=0):
  print("Leap Year")
else:
  print("Not a Leap Year")
```
